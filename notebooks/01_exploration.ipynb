# Synthetic Invoice Exploration

A compact exploration of the generated SAP-style accounts receivable data.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

project_root = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
df = pd.read_csv(project_root / "data" / "synthetic_invoices.csv")
df.head()
```

```python
print(df.shape)
print(f"Late payment rate: {df['target_paid_late'].mean():.1%}")
df.describe(include="all").T
```

```python
risk_by_dispute = df.groupby("dispute_flag")["target_paid_late"].mean()
risk_by_dispute.plot(kind="bar", title="Late-payment rate by dispute status")
plt.ylabel("Late-payment rate")
plt.show()
```

```python
df.groupby("dunning_level")["target_paid_late"].mean().plot(
    marker="o", title="Late-payment rate by dunning level"
)
plt.ylabel("Late-payment rate")
plt.show()
```
